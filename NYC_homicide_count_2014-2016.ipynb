** Note that NYC's homicide counts are constantly smaller than UCR report.** 

** Also, since we have two separate dataset, one is for 2005-2016 and one is for first queater in 2017, we have count incidents in 2017 (reported in 2017 but happened in 2016, and then add those numbers to 2016's counting.**

```python
import pandas as pd
from dateutil import parser
```

```python
nyc = pd.read_csv('/Volumes/TRANSCEND/Violence Tracker/NYC/NYPD_Crimes_2006-2015.csv')
```

```python
nyc.dtypes
```

```python
nyc.head() 
```

```python
'''
Note: it's gonna take a while to convert string to datetime for this 'large' dataframe, around 15min
'''
nyc['CMPLNT_FR_DT'] = pd.to_datetime(nyc['CMPLNT_FR_DT'],errors='coerce') 
```

```python
nyc.CMPLNT_FR_DT.dt.year.unique()
```

```python
nyc['Year'], nyc['Month'] = nyc['CMPLNT_FR_DT'].dt.year, nyc['CMPLNT_FR_DT'].dt.month
```

```python
'''
NYC's data from year 2014-2016
'''
nyc_14_16 = nyc[nyc['Year'] >= 2014]
```

```python
nyc.OFNS_DESC.unique()
```

```python
'''
Here we only count 'MURDER & NON-NEGL. MANSLAUGHTER' type 
Neg-Vehicle and HOMICIDE-NEGLIGENT does not included.
'''
#murder = ['MURDER & NON-NEGL. MANSLAUGHTER', 'HOMICIDE-NEGLIGENT-VEHICLE','HOMICIDE-NEGLIGENT,UNCLASSIFIE']
murder = ['MURDER & NON-NEGL. MANSLAUGHTER']
flag = nyc_14_16.OFNS_DESC.isin(murder)
nyc_count = nyc_14_16[flag].groupby(['Year','Month','OFNS_DESC']).size().reset_index(name='count')
```

```python
nyc_count.insert(0,'City','New York City')
nyc_count.insert(1,'State','NY')
nyc_count['UCR_Type'] = ''
nyc_count = nyc_count.rename(columns = {'OFNS_DESC':'homicide_type'})
```

Note that in 2017's dataset, there might be situations that crime report date is in 2017 but occurred date is in 2016, thus we need to add those number from 17's dataset.

```python
'''
Additional homicides counts from 2017's dataset (occurred in 2016 but reported in 2017) 
'''
addition_homi_for_16 = pd.read_csv("addition_homi_for_16.csv")
```

```python
addition_homi_for_16
```

```python
nyc_count
```

```python
'''
Add homicide incidents which recorded in 2017's dataset
'''
nyc_count = pd.concat([nyc_count,addition_homi_for_16]).\
groupby(['City','State','Year','Month','homicide_type']).sum().reset_index()
```

```python
nyc_14_count = nyc_count[nyc_count.Year==2014]['count'].sum()
nyc_15_count = nyc_count[nyc_count.Year==2015]['count'].sum()
nyc_16_count = nyc_count[nyc_count.Year==2016]['count'].sum()

print('Year 2014 homicide count',nyc_14_count )
print('Year 2015 homicide count',nyc_15_count )
print('Year 2016 homicide count',nyc_16_count )
```

Note that the result is constantly smaller than the offical UCR report for year 2014 and year 2015. UCR has homicide count **333** for 2014 and **352** for 2015. Please refer to NYPD's official published report checking for consistency.
