```python
import pandas as pd
from dateutil import parser
```

```python
omada_2014 = pd.read_excel("/Volumes/TRANSCEND/Violence Tracker/Omaha, NE/Omaha_NE_Homicide_Data_2014.xlsx")
omada_2015 = pd.read_excel("/Volumes/TRANSCEND/Violence Tracker/Omaha, NE/Omaha_NE_Homicide_Data_2015.xlsx")
omada_2016 = pd.read_excel("/Volumes/TRANSCEND/Violence Tracker/Omaha, NE/Omaha_NE_Homicide_Data_2016.xlsx")
omada_2017 = pd.read_excel("/Volumes/TRANSCEND/Violence Tracker/Omaha, NE/Omaha_NE_Homicide_Data_2017.xlsx")
```

```python
omada_2014.head() 
```

```python
#concatenate omada's data from 2014-2017
frames = [omada_2014,omada_2015,omada_2016,omada_2017]
df = pd.concat(frames);df.head()
```

```python
#dt = parser.parse(df['Date'])
df['Date']=df['Date'].apply(parser.parse);df 
```

```python
df['year'], df['month'] = df['Date'].dt.year, df['Date'].dt.month
df.head()
```

```python
df_count = df.groupby(['year','month','Cause']).size().reset_index(name='count')
```

```python
df_count=df_count.rename(columns = {'Cause':'homicide_type'});df_count.head()
```

```python
df_count.insert(0,'City','Omaha') 
```

```python
df_count.insert(1,'State','NE')
```

```python
df_count['UCR_Type'] = ''
```

```python
df_count.head(15)
```

```python
df_count.to_csv('Omaha_NE_Homicide_Data_counts.csv',index=False,encoding='utf-8')
```
