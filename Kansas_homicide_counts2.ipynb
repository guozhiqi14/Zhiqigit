```python
def get_homicide_count(df):
    df['Reported_Date']=df['Reported_Date'].apply(parser.parse);
    df['year'], df['month'] = df['Reported_Date'].dt.year, df['Reported_Date'].dt.month
    
    df_homi = df[(df['Description'] == 'Justifiable Homicide') | (df['Description'] == 'HOMICIDE/Non Neglige')] 
    df_homi = df_homi.groupby(['year','month','Description','Involvement']).size().reset_index(name='count')
    
    df_count = df_homi[df_homi['Involvement'] == 'VIC'] ##count only victim
    df_count.insert(0,'City','Kansas')
    df_count.insert(1,'State','MO')
    df_count['UCR_Type'] = ''
    
    return df_count
```

```python
df_count_2014 = get_homicide_count(df2014) 
```

```python
df_count_2015 = get_homicide_count(df2015)
df_count_2016 = get_homicide_count(df2016)
df_count_2017 = get_homicide_count(df2017) 
```

```python
frame = [df_count_2014,df_count_2015,df_count_2016,df_count_2017] 
df_final = pd.concat(frame)
```

```python
df_final
#df_final = df_final.drop('Involvement', 1)
df_final=df_final.rename(columns = {'Description':'homicide_type'})
```
