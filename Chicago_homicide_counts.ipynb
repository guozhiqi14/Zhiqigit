```python
import pandas as pd
from dateutil import parser
```

```python
chicago = pd.read_csv('/Volumes/TRANSCEND/Violence Tracker/Chicago/CPD_Crimes_2001_to_present.csv')
```

```python
chicago
```

```python
chicago.dtypes
```

```python
chicago['Primary Type'].unique()
```

```python
#chicago
chicago['Date'] = pd.to_datetime(chicago['Date'])
```

```python
chicago['year'],chicago['month'] = chicago['Date'].dt.year, chicago['Date'].dt.month
```

```python
chicago.head()
```

```python
chicago_14_17 = chicago[chicago['year'] > 2013]
```

```python
chicago_homicide = chicago_14_17[chicago_14_17['Primary Type'] == 'HOMICIDE']
```

```python
chicago_homicide.head()
```

```python
chicago_count = chicago_homicide.groupby(['year','month','Description']).size().reset_index(name='count');chicago_count
```

```python
chicago_count.insert(0,'City','Chicago')
chicago_count.insert(1,'State','IL')
chicago_count['UCR_Type'] = ''
chicago_count.head(15)
```

```python
chicago_count.to_csv('CPD_Crimes_2001_to_present_homicide_counts.csv',index=False,encoding='utf-8')
```
