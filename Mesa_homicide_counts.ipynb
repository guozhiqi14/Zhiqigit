```python
import pandas as pd
from dateutil import parser
import time
```

```python
df = pd.read_csv('/Volumes/TRANSCEND/Violence Tracker/Mesa, AZ/Mesa_Police_Crime_Data.csv')
```

```python
df.shape
```

```python
df['REPORT_DATE'] = pd.to_datetime(df['REPORT_DATE'])
```

```python
df['year'], df['month'] = df['REPORT_DATE'].dt.year, df['REPORT_DATE'].dt.month
```

```python
df['INCIDENT_TYPE'].unique()
```

```python
#Did not count NEGLIGENT HOMICIDE
murder_grp = ['MURDER 2ND DEG-INTENTIONAL','MURDER 2ND DEG-INTENTIONAL-DV','MURDER 1ST DEG-PREMEDITATED',
              'MURDER 1ST DEG-DURING CRIME','MURDER 2ND DEGREE - EXT INDIFFERENCE','MURDER 2ND DEGREE',
              'MURDER 1ST DEGREE','MURDER 2ND DEG-KNOWING']
```

```python
murder = df['INCIDENT_TYPE'].isin(murder_grp)
```

```python
df[murder].shape
```

```python
df_count = df[murder].groupby(['year','month','INCIDENT_TYPE']).size().reset_index(name = 'count')
```

```python
df_count=df_count.rename(columns = {'INCIDENT_TYPE':'homicide_type'})
df_count.insert(0,'City','Mesa')
df_count.insert(1,'State','AZ')
df_count['UCR_Type'] = ''
```

```python
df_count
```

```python
df_count.to_csv('Mesa_Police_Crime_Data_homicide_counts.csv',index=False,encoding='utf-8')
```
