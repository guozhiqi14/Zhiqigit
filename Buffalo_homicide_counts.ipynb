```python
import pandas as pd
from dateutil import parser
```

```python
'''
Read homicide listings for Buffalo City
'''
Buffalo_14 = pd.read_excel('/Volumes/TRANSCEND/Violence Tracker/Buffalo, NY/Buffalo_city_homicide_data_2014.xlsx')
Buffalo_15 = pd.read_excel('/Volumes/TRANSCEND/Violence Tracker/Buffalo, NY/Buffalo_city_homicide_data_2015.xlsx')
Buffalo_16 = pd.read_excel('/Volumes/TRANSCEND/Violence Tracker/Buffalo, NY/Buffalo_city_homicide_data_2016.xlsx')
```

```python
Buffalo_14['Date'] = pd.to_datetime(Buffalo_14['Date']) 
```

```python
#Buffalo_14.dtypes
Buffalo_14['Year'], Buffalo_14['Month'] = Buffalo_14['Date'].dt.year, Buffalo_14['Date'].dt.month 
```

```python
Buffalo_homi_count_14 = Buffalo_14.groupby(['Year','Month','Status']).size().reset_index(name='count')
```

```python
Buffalo_homi_count_14.insert(0,'City','Buffalo')
Buffalo_homi_count_14.insert(1,'State','NY')
Buffalo_homi_count_14['UCR_Type'] = ''
```

```python
Buffalo_homi_count_14.to_csv('Buffalo_city_homicide_count_2014.csv',index=False,encoding='utf-8')
```

```python
def get_homi_count():
    frame = [Buffalo_15,Buffalo_16] 
    df = pd.concat(frame,ignore_index=True)
    df['Year'], df['Month'] = df['Date'].dt.year, df['Date'].dt.month 
    df_count = df.groupby(['Year','Month','Method']).size().reset_index(name='count')
    df_count.insert(0,'City','Buffalo')
    df_count.insert(1,'State','NY')
    df_count['UCR_Type'] = ''
    df_count = df_count.rename(columns = {'Method':'homicide_type'})
    
    return df_count
```

```python
#Count for year 2015 and 2016
Buffalo_homi_count_15_16 = get_homi_count() 
```

```python
Buffalo_homi_count_15_16.to_csv('Buffalo_city_homicide_count_2015-2016.csv',index=False,encoding='utf-8')
```
