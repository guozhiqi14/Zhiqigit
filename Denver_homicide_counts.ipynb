```python
denver['year'], denver['month'] = denver['FIRST_OCCURRENCE_DATE'].dt.year, denver['FIRST_OCCURRENCE_DATE'].dt.month
```

```python
denver_count = denver[(denver['year'] > 2013) & (denver['OFFENSE_CATEGORY_ID']=='murder')]
```

```python
denver_count['OFFENSE_TYPE_ID'].unique()
```

```python
denver_count2=denver_count.groupby(['year','month','OFFENSE_TYPE_ID']).size().reset_index(name='count') 
```

```python
#denver_count2
denver_count2.insert(0,'City','Denver')
denver_count2.insert(1,'State','CO')
denver_count2['UCR_Type'] = ''
denver_count2
```

```python
denver_count2.to_csv('Denver_DPD_Crime_Data_2012-2017_homicide_counts.csv',index=False,encoding='utf-8')
```
