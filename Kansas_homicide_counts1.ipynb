```python
frames = [df2014,df2015,df2016,df2017]
df = pd.concat(frames);df.head()
```

```python
df.columns
```

```python
df['Reported_Date']=df['Reported_Date'].apply(parser.parse);
```

```python
df['year'], df['month'] = df['Reported_Date'].dt.year, df['Reported_Date'].dt.month
df.head()
```

```python
df['Description'].unique()
```

```python
df = df[(df['Description'] == 'Justifiable Homicide') | (df['Description'] == 'HOMICIDE/Non Neglige')]
```

```python
df_count = df.groupby(['year','month','Description']).size().reset_index(name='count')
```

```python
df_count=df_count.rename(columns = {'Description':'homicide_type'});df_count.head()
```

```python
df_count.insert(0,'City','Kansas')
df_count.insert(1,'State','MO')
df_count['UCR_Type'] = ''
df_count.head(15)
```
