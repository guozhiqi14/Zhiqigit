```python
df['incident_datetime']=df['incident_datetime'].apply(parser.parse);
```

```python
df['year'], df['month'] = df['incident_datetime'].dt.year, df['incident_datetime'].dt.month 
```

```python
df['year'].unique()
```

```python
a = df[df['year'] > 2013]
```

Computer-Aided Dispatch (**CAD**) and automated Records Management Systems (**RMS**) are powerful tools that support the day-to-day operations of law enforcement agencies, enhance public safety, and support efficient use of resources and effective tactical deployment. Moreover, CAD and RMS systems are also critical tools that law enforcement agencies use to support crime analysis, community policing, and information sharing programs. 

```python
b = a[(a['incident_type_primary'] == 'CAD: HOMICIDE') | (a['incident_type_primary'] == 'RMS: MURDER & NON NEG HOMICIDE')]
```

```python
c = b.groupby(['year','month','incident_type_primary']).size().reset_index(name='count') 
```

```python
c=c.rename(columns = {'incident_type_primary':'homicide_type'})
c.insert(0,'City','Fresno')
c.insert(1,'State','CA')
c['UCR_Type'] = ''
```

```python
c
```

```python
c.to_csv('Fresno_Police_Department_Crime_Data_homicide_counts.csv',index=False,encoding='utf-8')
```

```python
#Murder instances at Fresno in 2014
c[c.year == 2014]['count'].sum()
```
